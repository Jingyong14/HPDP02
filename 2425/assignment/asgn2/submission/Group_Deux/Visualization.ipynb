```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set(style="whitegrid")

data = {
    'Method': [
        'Pandas (No Optimization)',
        'Column Selection',
        'Chunking',
        'Data Type Optimization',
        'tratified Sampling',
        'Dask',
        'Polars'
    ],
    'Execution Time (s)': [82.33, 48.43, 0.59, 84.25, 37.04, 0.07, 11.98],
    'Memory Usage (MB)': [9062.15, 2133.99, 60.63, 8955.10, 911.93, 0, 2528.16]
}

df = pd.DataFrame(data)

# Plot 1: Execution Time - Line Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Method', y='Execution Time (s)', data=df, marker='o', sort=False, color='blue')
plt.title('Execution Time Comparison (Line Plot)')
plt.ylabel('Execution Time (seconds)')
plt.xlabel('Method')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot 2: Memory Usage - Horizontal Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(y='Method', x='Memory Usage (MB)', data=df, palette='Reds_d')
plt.title('Memory Usage Comparison (Horizontal Bar Plot)')
plt.xlabel('Memory Usage (MB)')
plt.ylabel('Method')
plt.tight_layout()
plt.show()
```
