```python
# Time of processing for each library (default, Polars, and Dask)
time_def = [27.03, 25.92, 27.54]
time_polars = [11.09, 12.62, 16.10]
time_dask = [41.85, 42.19, 42.55]
```

```python
# Memory used for each library (default, Polars, and Dask)
mem_def = [1347.09, 1396.53, 1347.01]
mem_polars = [1124.07, 1127.49, 1050.79]
mem_dask = [34.6, 21.10, 13.61]
```

```python
import matplotlib.pyplot as plt
import numpy as np

runs = [1, 2, 3]
x = np.arange(len(runs))  # [0,1,2]

width = 0.25  # width of the bars

# Bar Chart for Execution Time
fig, ax = plt.subplots(figsize=(10,5))
bars1 = ax.bar(x - width, time_def, width, label='Pandas')
bars2 = ax.bar(x, time_polars, width, label='Polars')
bars3 = ax.bar(x + width, time_dask, width, label='Dask')

ax.set_xlabel('Run Number')
ax.set_ylabel('Time (seconds)')
ax.set_title('Processing Time Comparison')
ax.set_xticks(x)
ax.set_xticklabels(runs)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

fig.savefig("processing_time.png", dpi=300)
```

```python
# Line Chart for Memory Usage
runs = [1, 2, 3]

fig2, ax2 = plt.subplots(figsize=(10,5))

ax2.plot(runs, mem_def, marker='o', label='Pandas')
ax2.plot(runs, mem_polars, marker='s', label='Polars')
ax2.plot(runs, mem_dask, marker='^', label='Dask')

ax2.set_xlabel('Run Number')
ax2.set_ylabel('Memory Used (MB)')
ax2.set_title('Memory Usage Comparison')
ax2.legend()
ax2.set_xticks(runs)
ax2.grid(True, linestyle='--', alpha=0.7)

plt.show()

fig2.savefig("memory_usage.png", dpi=300)
```
