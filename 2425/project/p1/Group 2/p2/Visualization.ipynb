```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['Without Optimization', 'Polars', 'Modin & Ray', 'Multiprocessing', 'Joblib', 'Spark']
times = [109.48, 5.15, 26.41, 19.57, 20.19, 36.92]

# Create bar plot
plt.figure(figsize=(10, 8))
bars = plt.bar(methods, times, color='skyblue', edgecolor='black')

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.2f}',
             ha='center', va='bottom')

# Customize plot
plt.title('Processing Time Comparison of Optimization Methods', fontsize=14, pad=20)
plt.xlabel('Optimization Method', fontsize=12)
plt.ylabel('Processing Time (seconds)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the plot
plt.savefig('optimization_comparison.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['Without Optimization', 'Polars', 'Modin & Ray', 'Multiprocessing', 'Joblib', 'Spark']
memory_usage = [1242.47, 746.41, 393.36, 456.23, 567.96, 106.51]

# Create horizontal bar plot
plt.figure(figsize=(20, 6))
bars = plt.barh(methods, memory_usage, color='lightcoral', edgecolor='black')

# Add value labels at the end of each bar
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + 20, bar.get_y() + bar.get_height()/2, f'{xval:.2f}',
             va='center', ha='left')

# Customize plot
plt.title('Memory Usage Comparison of Optimization Methods', fontsize=14, pad=20)
plt.xlabel('Memory Usage (MB)', fontsize=12)
plt.ylabel('Optimization Method', fontsize=12)
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot
plt.savefig('memory_usage_comparison.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['Without Optimization', 'Polars', 'Modin & Ray', 'Multiprocessing', 'Joblib', 'Spark']
peak_memory = [460.24, 99.14, 71.74, 107.99, 385.68, 0.15]

# Create line plot with markers
plt.figure(figsize=(10, 12))
plt.plot(methods, peak_memory, marker='o', color='teal', linestyle='-', linewidth=2, markersize=8)

# Add value labels above each point
for i, usage in enumerate(peak_memory):
    plt.text(i, usage + 15, f'{usage:.2f}', ha='center', va='bottom')

# Customize plot
plt.title('Peak Memory Usage Comparison of Optimization Methods', fontsize=14, pad=20)
plt.xlabel('Optimization Method', fontsize=12)
plt.ylabel('Peak Memory Usage (MB)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the plot
plt.savefig('peak_memory_usage_comparison.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['Without Optimization', 'Polars', 'Modin & Ray', 'Multiprocessing', 'Joblib', 'Spark']
throughput = [1318.09, 27990.23, 5460.38, 7368.01, 7141.52, 3906.31]

# Create line plot
plt.figure(figsize=(10, 16))
plt.plot(methods, throughput, color='purple', marker='o', markersize=8, linewidth=2)

# Add value labels above each point
for i, value in enumerate(throughput):
    plt.text(i, value + 1000, f'{value:.2f}', ha='center', va='bottom')

# Customize plot
plt.title('Throughput Comparison of Optimization Methods', fontsize=14, pad=20)
plt.xlabel('Optimization Method', fontsize=12)
plt.ylabel('Throughput (records/second)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the plot
plt.savefig('throughput_comparison.png')
```
