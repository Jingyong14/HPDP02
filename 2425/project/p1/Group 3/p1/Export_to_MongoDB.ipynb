# Libraries

```python
import pandas as pd
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default

# Authenticate
creds, _ = default()
gc = gspread.authorize(creds)
```

# Loading Data

### Nasional

```python
spreadsheet = gc.open('Nasional')

worksheet = spreadsheet.sheet1


# Get all values
rows = worksheet.get_all_values()

# Convert to DataFrame (first row as header)
nasional = pd.DataFrame(rows[1:], columns=rows[0])

nasional.head()
```

```python
nasional.info()
```

```python
nasional.drop('Category', axis=1, inplace=True)
nasional.rename(columns={'Tab': 'Category'}, inplace=True)
nasional = nasional[['Title', 'Category', 'Date', 'URL']]
nasional.head()
```

### Ekonomi, Ringgit, Luar Negara

```python
spreadsheet = gc.open('Ekonomi_Ringgit_LuarNegara')

worksheet = spreadsheet.sheet1


# Get all values
rows = worksheet.get_all_values()

# Convert to DataFrame (first row as header)
erl = pd.DataFrame(rows[1:], columns=rows[0])

erl.head()
```

```python
erl.info()
```

```python
erl = erl[['Title', 'Category', 'Date', 'URL']]
erl.head()
```

### Sukan, Rencana, Pancaindera

```python
spreadsheet = gc.open('Sukan_Rencana_Pancaindera')

worksheet = spreadsheet.sheet1


# Get all values
rows = worksheet.get_all_values()

# Convert to DataFrame (first row as header)
srp = pd.DataFrame(rows[1:], columns=rows[0])

srp.head()
```

```python
srp.info()
```

```python
srp.drop('Category', axis=1, inplace=True)
srp.rename(columns={'Subcategory': 'Category', 'Link':'URL', 'DateTime':'Date'}, inplace=True)
srp = srp[['Title', 'Category', 'Date', 'URL']]
srp.head()
```

### Gaya, Komuniti

```python
spreadsheet = gc.open('Gaya_Komuniti')

worksheet = spreadsheet.sheet1


# Get all values
rows = worksheet.get_all_values()

# Convert to DataFrame (first row as header)
gk = pd.DataFrame(rows[1:], columns=rows[0])

gk.head()
```

```python
gk.info()
```

```python
gk.drop('Category', axis=1, inplace=True)
gk.rename(columns={'Subcategory': 'Category'}, inplace=True)
gk = gk[['Title', 'Category', 'Date', 'URL']]
gk.head()
```

### Total Data

```python
print("Total number of rows: ",len(nasional)+len(erl)+len(srp)+len(gk))
```

# Combining Datasets

```python
utusan_articles = pd.concat([nasional, erl, srp, gk], ignore_index=True)
utusan_articles
```

```python
utusan_articles.info()
```
